```python
import numpy as np
w_st = -100
w_end = 100
w = np.linspace(w_st, w_end, 1000)
x = [1, 2, 3]
y = [1, 2, 3]
f = np.zeros((len(w)*len(x))).reshape(len(w),len(x))
J = np.zeros(len(w))
m = len(x)
for i in range(len(w)):
    for j in range(len(x)):
        f[i,j] = w[i]*x[j]

for i in range(len(w)):
    for j in range(len(y)):
        J[i] = ((w[i]*x[j] - y[j])**2)*(1/(2*m))

```

```python
import matplotlib.pyplot as plt
plt.figure(num = 0, dpi = 120)
for i in range(len(J)):
    if J[i] == min(J):
        min_pos = i
        break
    else:
        continue
plt.plot(w,J)
plt.grid()
plt.scatter(w[min_pos], min(J))
plt.ylim(-500,15e03)

plt.figure(num = 1, dpi = 120)
plt.grid()
plt.plot(x, f[min_pos,:])
plt.scatter(x,y)
```
