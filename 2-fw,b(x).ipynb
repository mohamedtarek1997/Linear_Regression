```python
# 1-Import libraries 
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
```

```python
# 2-Read and show the csv file
df = pd.read_csv('movies.csv')
print(df.shape)
df.head(10)
```

```python
# 3-Drop rows with NAN's and duplicates and remember:
# axis = 0  is for rows 
# axis = 1 is for columns
nan = df[df.isna().any(axis=1)]
print(nan.shape)
nan.head()
```

```python
print(df.isna().sum())
print('data frame size before dropping nans and duplicates', df.shape)
df = df.dropna(axis = 0, how = 'any')
df = df.drop_duplicates()
df.shape
```

```python
# 4- We need to have a look on data types through our columns to see if we need to make some changes
print(df.dtypes)
```

```python
df['budget'] = df['budget'].astype('int')
df['gross'] = df['gross'].astype('int')
df['votes'] = df['votes'].astype('int')
df
```

```python
# 5- Sorting our data based on gross to see if there are negative values or any issues
df = df.sort_values( by = 'gross', ascending= False)
df
```

```python
x = list()
for i in df['gross']:
    i = abs(i)
    x.append(i)
df['gross'] = x
df = df.sort_values(by = 'gross', ascending=False)
df
```

```python
# Repeat the process for budget
df = df.sort_values(by = 'budget', ascending=False)
df
```

```python
# 6- Making a scatter plot of our variables for linear regression which are x = budget, y = gross
plt.figure(num = 0, dpi = 120)
plt.grid()
plt.scatter(df['budget'], df['gross'])
```

```python
# 7- Convert our target columns from being data frames to be arrays
x = df['budget'].to_numpy()
y = df['gross'].to_numpy()
```

# Now let's start with our linear regression model

```python
# Linear regression with unscaled values of x and y 
w = 0
b = 0
m = len(x)
alpha = 1e-17
itr_num = 300
J = list()
itr = list()

for i in range(itr_num):
    itr.append(i)
    j = ( 1/( 2*m ) ) * sum( ( ( w*x+b ) - y ) **2  )
    w2 = w - (  alpha * (1/m) * sum( ( ( w*x+b ) - y ) * x )  )
    b2 = b - (  alpha * (1/m) * sum( ( ( w*x+b ) - y )  )  ) 
    w = w2
    b = b2
    J.append(j)
    if j > 1e100:
        break

plt.figure(num = 1, dpi = 120)
plt.plot(itr, J)
plt.xlabel('# Iterations')
plt.ylabel('Loss Values')
plt.title('# Iterations VS Loss for Unscaled Values')


```

```python
plt.figure(num = 2, dpi = 120)
plt.scatter(x,y)
plt.plot(x, w*x+b, color='red')
plt.xlabel('Budget')
plt.ylabel('gross/revenue')
plt.title('money spent VS Revenues for Unscaled Values')
plt.grid()
```

```python
# Linear regression with scaled values of x and y 
x_scaled = x/max(x)
y_scaled = y/max(y)
w = 0
b = 0
m = len(x)
alpha = 0.1
itr_num = 3000
J = list()
itr = list()

for i in range(itr_num):
    itr.append(i)
    j = ( 1/( 2*m ) ) * sum( ( ( w*x_scaled+b ) - y_scaled ) **2  )
    w2 = w - (  alpha * (1/m) * sum( ( ( w*x_scaled+b ) - y_scaled ) * x_scaled )  )
    b2 = b - (  alpha * (1/m) * sum( ( ( w*x_scaled+b ) - y_scaled )  )  ) 
    w = w2
    b = b2
    J.append(j)
    if j > 1e100:
        break

plt.figure(num = 3, dpi = 120)
plt.plot(itr, J)
plt.xlabel(' Iterations ')
plt.ylabel(' Loss ')
plt.title(' Iterations VS Loss for Scaled Values ')


```

```python
plt.figure(num = 4, dpi = 120)
plt.scatter(x,y)
plt.plot(x, (w*x_scaled+b)*max(y), color='red')
plt.xlabel('Budget')
plt.ylabel('gross/revenue')
plt.title('money spent VS Revenues')
plt.grid()
```

# Another awesome approach using matrix multiplication and pseudoinverse

```python
xm = np.matrix(x)
xm.shape
```

```python
ym = np.matrix(y)
ym.shape
```

```python
m, n = xm.shape
un = np.matrix(np.ones((m, n)))
un
```

```python
X = np.concatenate((un, xm), axis=0)
X
```

```python
X.T
```

<img src="Inv2 EX.png" style="width:700px; height:900">

<img src="Inv2 EX2.png" style="width:700px; height:900">

```python
W = np.dot(np.linalg.pinv(X).T, y.T)
```

```python
W
```

```python
plt.figure(num = 4, dpi = 130)
plt.scatter(x, y)
plt.grid()
plt.plot(x, W[0, 1]*x+W[0, 0], 'red')
plt.title('           Budget VS Gross Using Pseudoinvers "AWESOOOOOOOOME"')
plt.xlabel('Budget')
plt.ylabel('Gross')
plt.xlim(0.0, 3.71e08)
```
